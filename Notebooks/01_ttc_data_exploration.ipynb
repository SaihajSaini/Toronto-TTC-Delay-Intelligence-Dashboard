```python
clean_df = df.copy()
```

```python
clean_df = clean_df.drop(columns=["_id"])
```

```python
clean_df = clean_df.rename(columns={
    "Min Delay": "Min_Delay",
    "Min Gap": "Min_Gap"
})
```

```python
clean_df["Date"] = pd.to_datetime(clean_df["Date"])
```

```python
clean_df["Bound"] = clean_df["Bound"].fillna("Unknown")

clean_df["Line"] = clean_df["Line"].fillna("Unknown")
```

```python
clean_df.info()
```

```python
clean_df.isnull().sum()
```

```python
clean_df["Month"] = clean_df["Date"].dt.month_name()
```

```python
monthly_delays = clean_df["Month"].value_counts()

monthly_delays
```

```python
month_order = [
    "January",
    "February",
    "March",
    "April",
    "May",
    "June",
    "July",
    "August",
    "September",
    "October",
    "November",
    "December"
]

monthly_delays = monthly_delays.reindex(month_order)

monthly_delays
```

```python
clean_df.to_csv("../data/processed/ttc_delays_cleaned.csv", index=False)
```
